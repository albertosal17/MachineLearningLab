```python
import matplotlib.pyplot as plt
import csv
from scipy import stats
import numpy as np
import sklearn as sl
```

```python
# Load the provided data file with the used car data (you can also have a look at it with any text editor)

filename = "data/km_year_power_price.csv"
lines = csv.reader(open(filename, newline=''), delimiter=',') #ogni linea di lines è una lista

# place your loading code here
#print(lines) #lines non è una lista!
dataset = list(lines) #converto lines in una lista (di liste)
dataset.pop(0) #rimuovo i nomi per le classi della tabella
#print(dataset) 
```

```python
#converto i dati riguardo la singola macchina in float
for i in range(len(dataset)):
    dataset[i] = [float(x) for x in dataset[i]]

dataset=np.array(dataset) #per poter fare la moltiplicazione fra numpy devo convertire la lista in un numpy array altrimenti da errore (liste e array sono leggermente diverse)
#print(dataset)
```

Use linear regression to estimate the car prices from the year, kilometers or engine power. 
You can make a simple 1D regression from each one of the parameters independently 




```python
# linear regression with linregress (estimate price from year)

years=dataset[:,1] # una lista con tutti gli elementi 1 (cioè gli anni) di tutte le righe
prices=dataset[:,3]

res_years = stats.linregress(years, prices)

#plotto i risultati
plt.plot(years, prices, '+', label='original data', markersize=5)
plt.xlabel('Year')
plt.ylabel('Price')

plt.plot(years, res_years.intercept + res_years.slope*years, 'r', label='fitted line')

plt.legend()

plt.show()
```

```python
# linear regression with linregress (estimate price from power)

power=dataset[:,2]

res_power = stats.linregress(power, prices)

#plotto i risultati
plt.plot(power, prices, '+', label='original data', markersize=5)
plt.xlabel('Power')
plt.ylabel('Price')

plt.plot(power, res_power.intercept + res_power.slope*power, 'r', label='fitted line')

#plt.legend()

plt.show()
```

```python
# linear regression with linregress (estimate price from km)

km=dataset[:,0] 

res_km = stats.linregress(km, prices)

#plotto i risultati
plt.plot(km, prices, '+', label='original data', markersize=5)
plt.xlabel('km')
plt.ylabel('Price')

plt.plot(km, res_km.intercept + res_km.slope*km, 'r', label='fitted line')

#plt.legend()

plt.show()
```

```python
# Have a look at the correlation coefficients to see which of the 3 features works better

print('years vs. prices', res_years.rvalue)
print('power vs. prices', res_power.rvalue)
print('km vs. prices', res_km.rvalue)

```
